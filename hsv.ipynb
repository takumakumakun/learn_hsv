```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def bgr_show(bgr):
    rgb = bgr[:,:,[2,1,0]]
    plt.axis('off')
    plt.imshow(rgb)

def bgr_show_side(bgr1, bgr2):
    rgb1 = bgr1[:,:,[2,1,0]]
    rgb2 = bgr2[:,:,[2,1,0]]
    plt.subplot(1,2,1)
    plt.axis('off')
    plt.imshow(rgb1)
    plt.subplot(1,2,2)
    plt.axis('off')
    plt.imshow(rgb2)
```

```python
img_bgr = cv2.imread("./len_std.jpg")
bgr_show(img_bgr)
```

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
img_hsv2 = np.copy(img_hsv)


# (Hue)[0,179]
img_hsv2[:,:,0] = img_hsv[:,:,0] + 10

# (Saturation)[0,255]
img_hsv2[:,:,1] = img_hsv[:,:,1] + 10

# (Value)[0,255]
img_hsv2[:,:,2] = img_hsv[:,:,2] - 50


img_bgr2 = cv2.cvtColor(img_hsv2, cv2.COLOR_HSV2BGR)
bgr_show_side(img_bgr, img_bgr2)
```
